```
import numpy as np
import pylab as pl
#!pip install astropy
import astropy
from astropy import units as u


%pylab inline
```

```
def square(l, ax=ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
  
```

```
def circle(l, ax=ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
  
```

```
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
```

```
def isin(x, y, l):
  return np.sqrt(x**2 + y**2) < l 
```

```
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
N = 10000
inner = 0
for i in range(N):
  x = np.random.rand() * 2 - 1
  y = np.random.rand() * 2 - 1
  if isin(x, y, 1):
    inner += 1
    color = 'r'
  else: 
    color = 'k'
  ax.plot(x, y, '.', c=color, alpha=0.5)
4 * inner / N
```
